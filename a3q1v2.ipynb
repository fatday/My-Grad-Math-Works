```python
import numpy as np
```

```python

def f(x1, x2):
  return np.sin(np.pi * x1) * (np.sin(np.pi * x2) + np.sin(2 * np.pi * x2))
```

```python
def u(x1, x2):
  coef1, coef2 = 1 / (2 * (np.pi)**2), 1 / (5 * (np.pi)**2)
  base1, base2 = np.sin(np.pi * x1) * np.sin(np.pi * x2), np.sin(np.pi * x1) * np.sin(2 * np.pi * x2)
  return coef1 * base1 + coef2 * base2
```

```python
def Finite_Difference_Approx(h, space):
  Length = space[1] - space[0]
  M = int(Length / h)
  n = M - 1

  # Construct B and A
  B = (np.diag(2.0 * np.ones(n), 0)  + np.diag(-1.0 * np.ones(n-1), 1) + np.diag(-1.0 * np.ones(n-1), -1))
  I = np.eye(n)
  A = (np.kron(B, I) + np.kron(I, B)) / (h**2)
  #print(A)
  u_vec = np.zeros(n*n)
  for i in range(n):
    for j in range(n):
      x1, x2 = (i + 1) * h, (j + 1) * h # get x1, x2 for h, 2h,...., (M-1)h
      u_vec[j * n + i] = u(x1, x2)
  estimated_f = A @ u_vec
  estimated_f = estimated_f.reshape((n, n))
  return estimated_f
```

```python
def get_matrix_index(target_val, h):
  target_x1, target_x2 = target_val[0], target_val[1]
  i, j = int(target_x1 / h) - 1, int(target_x2 / h) - 1
  return i, j
```

# Calculate the Approximate Solution by Finite Difference Method with $h=\frac{1}{10},\frac{1}{20}$

```python
h1 = 1/10
h2 = 1/20
space = [0, 1]

# h = 1/10
fj1 = Finite_Difference_Approx(h1, space)

# h = 1/20
fj2 = Finite_Difference_Approx(h2, space)
```
